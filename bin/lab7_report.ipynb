```python
import matplotlib.pyplot as plot
import numpy as np

plot.figure(figsize=(15, 10))
def y(x, w, s):
    return 1 / (1 + abs(x - w) / abs(s))

w = 2
X = np.linspace(-8, 10)

plot.plot(X, y(X, w, 1), label = "s = 1")
plot.plot(X, y(X, w, 0.25), label = "s = 0.25")
plot.plot(X, y(X, w, 4), label = "s = 4")
plot.legend()
```
