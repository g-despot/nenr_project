```python
import matplotlib.pyplot as plot
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def fun(x, y):
    return (((x-1)**2 + (y+2)**2 - 5 * x * y + 3) * np.cos(x/5)**2)

x = np.linspace(-4, 4, 9)
y = np.linspace(-4, 4, 9)

X, Y = np.meshgrid(x, y)
Z = fun(X, Y)
fig = plot.figure(figsize=(20, 20))
ax = plot.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
def f(a, b, x):
    return 1 / (1 + np.exp(b * (x - a)))

file = open("rulesGR.txt", "r")
rules = [[float(x) for x in line.strip().split(" ")] for line in file.readlines()]
file.close()

figure = plot.figure(figsize=(20, 80))
X = np.linspace(-4, 4)

i = 1
for tmp, rule in enumerate(rules):
    a, b, c, d, p, q, r = rule
    
    plot.subplot(10, 2, i)
    plot.ylim(ymax = 1)
    plot.plot(X, f(a, b, X))
    i = i + 1

    plot.subplot(10, 2, i)
    plot.ylim(ymax = 1)
    plot.plot(X, f(c, d, X))
    i = i + 1
```
